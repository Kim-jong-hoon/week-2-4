Matplotlib 플로팅

x 및 y 점 표시
이 plot()함수는 다이어그램에 점(마커)을 그리는 데 사용됩니다.

기본적으로 이 plot()함수는 지점 간에 선을 그립니다.

이 함수는 다이어그램의 지점을 지정하기 위한 매개변수를 사용합니다.

매개변수 1은 x축의 점을 포함하는 배열입니다 .

매개변수 2는 y축의 점을 포함하는 배열입니다 .

(1, 3)에서 (8, 10)까지 선을 그려야 하는 경우 plot 함수에 두 개의 배열 [1, 8]과 [3, 10]을 전달해야 합니다.

ex) 다이어그램에서 위치 (1, 3)에서 위치 (8, 10)까지 선을 그립니다.


```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

다이어그램에 두 점을 그립니다. 하나는 위치 (1, 3)에, 다른 하나는 위치 (8, 10)에 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

다이어그램에서 위치 (1, 3)에서 (2, 8)까지, 그 다음 (6, 1)까지, 마지막으로 위치 (8, 10)까지 선을 그립니다.

제목과 레이블에 대한 글꼴 속성을 설정합니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

**Matplotlib 파이 차트**

---

Pyplot을 사용하면 다음 pie()함수를 사용하여 원형 차트를 그릴 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

매개변수를 사용하여 원형 차트에 레이블을 추가합니다 labels.

매개 labels변수는 각 쐐기마다 하나의 레이블이 있는 배열이어야 합니다.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

시작 각도

---


앞서 언급했듯이 기본 시작 각도는 x축에 있지만 startangle매개변수를 지정하여 시작 각도를 변경할 수 있습니다.

매개 startangle변수는 각도로 정의되며 기본 각도는 0입니다.


> ex 첫 번째 쐐기를 90도에서 시작하세요.




```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

웨지 중 하나를 눈에 띄게 하고 싶으신가요? explode매개변수를 사용하면 가능합니다.

매개 explode변수가 지정되면 None각 쐐기마다 하나의 값을 갖는 배열이어야 합니다.

각 값은 각 쐐기가 중심에서 얼마나 떨어져 표시되는지를 나타냅니다.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

그림자




shadows매개변수를 다음과 같이 설정하여 원형 차트에 그림자를 추가합니다 True.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

범례에 헤더를 추가하려면 함수 title에 매개변수를 추가하세요 legend .

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
